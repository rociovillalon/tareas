```python
# Importar bibliotecas necesarias
import pandas as pd
import altair as alt

# Cargar el archivo CSV
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Cadenas Nivel Nacional.csv')

# Strip leading/trailing spaces from column names
data.columns = data.columns.str.strip()

# Limpiar y procesar los datos
# Eliminar columnas que no aportan valor a la visualización
columnas_relevantes = ['Cadenas', 'Reviews', 'Ratings']
datos_filtrados = data[columnas_relevantes].dropna()

# Crear un gráfico de barras básico para visualizar Compliance Score por Control
chart = alt.Chart(datos_filtrados).mark_bar().encode(
    x=alt.X('Cadenas', sort='-y', title='Cadenas'),
    y=alt.Y('Reviews', title='Reviews'),
    tooltip=['Cadenas', 'Reviews']
).properties(
    title='Cantidad de Reviews de Cadenas a Nivel Nacional',
    width=600,
    height=400
)

chart
```

```python
# Importar bibliotecas necesarias
import pandas as pd
import altair as alt

# Cargar el archivo CSV
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Cadenas Nivel Nacional.csv')

# Strip leading/trailing spaces from column names
data.columns = data.columns.str.strip()

# Limpiar y procesar los datos
# Eliminar columnas que no aportan valor a la visualización
columnas_relevantes = ['Cadenas', 'Reviews', 'Ratings']
datos_filtrados = data[columnas_relevantes].dropna()

# Crear un gráfico de barras básico para visualizar Compliance Score por Control
chart = alt.Chart(datos_filtrados).mark_bar().encode(
    x=alt.X('Cadenas', sort='-y', title='Cadenas'),
    y=alt.Y('Reviews', title='Reviews'),
    tooltip=['Cadenas', 'Reviews']
).properties(
    title='Cantidad de Reviews de Cadenas a Nivel Nacional',
    width=600,
    height=400
)


# Guardar el gráfico como archivo HTML para visualizarlo en un navegador
chart.save('visualización1.html')

# Mostrar un mensaje indicando que el archivo ha sido creado
print("El gráfico se ha guardado como 'visualización1.html'. Ábrelo en un navegador para visualizarlo.")
```
